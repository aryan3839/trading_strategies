```python
def top_performers(date):
  all_=mtl_12.loc[date]
  top=all_.nlargest(5)
  relevent_ret=mtl_ret[top.name:][1:2][top.index]
  return (relevent_ret).mean(axis=1).values[0]
```

```python
top_performers('2012-12-31')
```

```python
mom_ret=[]
for date in mtl_12.index[:-1]:
  mom_ret.append(top_performers(date))
```

```python
def calculate_max_drawdown(pnl_list):
    """
    Calculates the maximum drawdown of a portfolio given a list of daily percentage returns.

    Args:
        pnl_list: A list of daily percentage returns (floats).

    Returns:
        The maximum drawdown as a float.
    """

    if not pnl_list:
        return 0.0

    max_value = pnl_list[0]
    max_drawdown = 0.0

    for pnl in pnl_list:
        max_value = max(max_value, pnl)
        max_drawdown = max(max_drawdown, max_value - pnl)

    return max_drawdown

# Example usage
pnl_list_aroundzero = [pnl-1 for pnl in pnl_list]
max_drawdown = calculate_max_drawdown(pnl_list_aroundzero)
print(f"Maximum drawdown: {max_drawdown:.2%}")
```

```python
# Calculate portfolio prices
corpus = 100
corpus_vals = []
for ret in mom_ret:
  corpus_vals.append(corpus*ret)
  corpus = (corpus * ret)
```

```python
import sys
```

```python
# Leetcode algorithm to calculate max drawdown
n = len(corpus_vals)
maxi = corpus_vals[0]
max_prof = -sys.maxsize - 1
max_drawdown = -sys.maxsize - 1

for i in range(n):
  maxi = max(corpus_vals[i], maxi);
  max_loss = max(max_prof, maxi - corpus_vals[i]);
  max_drawdown = max(max_drawdown, max_loss/maxi)
```

```python
print("Drawdown of the strategy is: {}%".format(max_drawdown*100))
```
