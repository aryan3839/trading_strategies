```python
import yfinance as yf
import pandas as pd
import ta
```

```python
df=yf.download('AAPL',start='2015-01-01')
```

```python
df
```

```python
df['SMA']=df['Close'].rolling(200).mean()
```

```python
df
```

```python
df['RSI']=ta.momentum.rsi(df['Close'],window=10)
```

```python
df
```

```python
df['Price']=df['Open'].shift(-1)
```

```python
df
```

```python
df['Buy']=(df['Close']>df['SMA'])&(df['RSI']<30)
```

```python
df
```

```python
df[df.Buy]
```

```python
df['Sell']=(df['RSI']>40)
```

```python
df
```

```python
in_position=False #this is just showing if we have open position or not
buys,sells=[],[]
buydates,selldates=[],[]
for index,row in df.iterrows():
  if not in_position:
    if row.Buy:
      buys.append(row.Price)
      buydates.append(index)
      in_position=True
      r_count=0
  if in_position:
    r_count+=1
    if row.Sell:
      sells.append(row.Price)
      selldates.append(index)
      in_position=False
    elif r_count>=10:
      sells.append(row.Price)
      selldates.append(index)
      in_position=False
```

```python
selldates
```

```python
trades=pd.DataFrame([buys,sells],index=['Buys','Sells'])
```

```python
trades
```

```python
trades=trades.T
```

```python
trades['PnL']=(trades['Sells']-trades['Buys'])/trades['Buys']
```

```python
trades
```

```python
trades[trades['PnL']>0].shape[0]/trades.shape[0]   #It is used to calculates winning percentage win rate
```

```python
(trades['PnL']+1).cumprod()-1
```
